```python
from sklearn.model_selection import RandomizedSearchCV
import decision_tree as dt
x_train, y_train, x_test, y_test = dt.split_cifar_data()
from sklearn.tree import DecisionTreeClassifier
```

```python
params = {
    'max_depth': [10, 20, 30],  # Árboles más profundos capturan más patrones
    'min_samples_split': [2, 5, 10],  # Valores bajos permiten más crecimiento
    'min_samples_leaf': [1, 2, 5],  # Evita hojas con muy pocos datos
    'criterion': ['gini', 'entropy']  # Prueba ambos criterios
}

clf = RandomizedSearchCV(DecisionTreeClassifier(), params, cv=5, scoring='accuracy', n_iter=20, verbose=3)
```

```python
clf.fit(x_train, y_train)
```

```python
print("Mejores hiperparámetros:", clf.best_params_)
```
